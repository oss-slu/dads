```python
#1.4
import numpy as np

import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

from sklearn.metrics import silhouette_score



data = np.array([[2,10],[2,5], [8,4], [5,8], [7,5], [6,4], [1,2], [4,9]])



elbow={"inertia":[],"clusters":[]}

silhouette={"silhouette":[],"clusters":[]}



# Try different number of clusters, and evaluate the clusters using silhousett

for i in range(2,7):

 print("Clustering data into ", i, ' groups')

 kmeans_model=KMeans(i, init='k-means++', max_iter=300, n_init=10, random_state=42)

 kmeans_model.fit(data)



 elbow["clusters"].append(i)

 elbow["inertia"].append(kmeans_model.inertia_)

 silhouette["silhouette"].append(silhouette_score(data,kmeans_model.labels_))



# Visualize the optimal clusters based on Inertia

plt.figure(figsize=(15,6))

plt.subplot(1,2,1)

plt.scatter(x=elbow["clusters"],y=elbow["inertia"])

plt.plot(elbow["clusters"], elbow["inertia"])

plt.xlabel("Number of Clusters", size=20)

plt.ylabel("Inertia", size=20)

plt.title("Elbow Graph", size=20)
```

```python
from sklearn.cluster import DBSCAN
import numpy as np
X = np.array([[2,10], [2,5], [8,4],
              [5,8], [7,5], [6,4], [1,2],[4,9]])
clustering = DBSCAN(eps=2, min_samples=3).fit(X)
clustering.labels_

plt.scatter(X[:, 0], X[:, 1], c=clustering.labels_)
plt.show()
```
